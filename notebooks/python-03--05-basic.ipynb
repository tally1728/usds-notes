### `*args`と`**kwargs`

`*args`はtuple型

```python
def print_list(*args):
    print(type(args))
    for elem in args:
        print(elem)
```

```python
print_list("java", "python", "c++")
```

`**kwargs`はdict型

```python
def print_dict(**kwargs):
    print(type(kwargs))
    print(kwargs)
```

```python
print_dict(key="value")
```

dict型だが、引数の取り出しは`kwargs[key]`でなく、`get()`メソッドを使う。

```python
def print_dict(**kwargs):
    param1 = kwargs.get("param1")
    param2 = kwargs.get("param2", "default_value")
    param3 = kwargs.get("param2")

    print("param1 is {}".format(str(param1)))
    print("param2 is {}".format(str(param2)))
    print("param3 is {}".format(str(param3)))
```

```python
print_dict(param1="param1_value")
```
