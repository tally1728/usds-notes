### 乱数生成

**補足**：NumPy1.17(2019/7)以降は`Generator`を利用する方法が推奨されている。

- `random.Generator.random(size=None, dtype=np.float64, out=None)`メソッド: 連続一様分布
- `random.Generator.normal(loc=0.0, scale=1.0, size=None)`メソッド: 正規分布
- `random.Generator.integers(low, high=None, size=None, dtype=np.int64, endpoint=False)`メソッド: 離散一様分布

```python
rng = np.random.default_rng(1728)
```

```python
rng.random()
```

```python
rng.random(3)
```

```python
rng.random((3, 5))
```

```python
rng.normal(size=3)
```

```python
rng.normal(size=(3, 4))
```

### random.Generator.integers(low, high=None, size=None, dtype=np.int64, endpoint=False)

```python
rng.integers(10, 50, (2, 4, 3))
```

### .reshape(shape)

```python
array = np.arange(0, 10)
print('shape:\n{}'.format(array.shape))
print('array:\n{}'.format(array))

reshaped_array = array.reshape((2, 5))
print('shape:\n{}'.format(reshaped_array.shape))
print('array:\n{}'.format(reshaped_array))

print('original array:\n{}'.format(array))
```

更に変形する。

```python
reshaped_array = reshaped_array.reshape((5, 2))
print('shape:\n{}'.format(reshaped_array.shape))
print('array:\n{}'.format(reshaped_array))
```
