## 20 matplotlibを使ってData Visualizationを始めよう

[データサイエンスのためのPython入門20〜matplotlibを使ってData Visualizationを始めよう〜 - 米国データサイエンティストのブログ](https://datawokagaku.com/data_vis_intro/)

### plt.plot()関数

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(-3, 3, 10)
y = np.exp(x)

print(x)
print(y)
```

```python
plt.plot(x, y)
```

#### スタイル
- marker: `^`, `o`
- line: `--`, 
- color: `g`, `r`, `b`

```python
plt.plot(x, y, "^g")
```

```python
plt.plot(x, y, "--")
```

```python
plt.plot(x, y, "r")
```

```python
plt.plot(x, y, "ob")
```

#### ラベル

```python
plt.plot(x, y)
plt.xlabel("Effort")
plt.ylabel("Earning")
plt.title("This is how your effor earns")
plt.show()
```

```python
plt.plot(x, y, label="Earning with effort")
plt.legend()
plt.xticks(np.arange(-3, 4, 0.5))
plt.yticks([0, 5, 10, 20])
plt.show()
```

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x) * 2

plt.plot(x, y1, label="first")
plt.plot(x, y2, label="second")
plt.legend()
plt.axis("off")
plt.show()
```

## 21 matplotlibのsubplotとオブジェクト指向を使って複数のグラフを並べる

[データサイエンスのためのPython入門21〜matplotlibのsubplotとオブジェクト指向を使って複数のグラフを並べる〜 - 米国データサイエンティストのブログ](https://datawokagaku.com/matplotlib_subplot/)

### plt.subplot(nrows, ncols, index)関数

注意：`index`は`1`から始まる

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = x**2

plt.subplot(1, 2, 1)
plt.plot(x, y1)

plt.subplot(1, 2, 2)
plt.plot(x, y2)

plt.show()
```

### オブジェクト指向による記述

```python
fig = plt.figure()
print(type(fig))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
print(type(ax1))
print(type(ax2))

ax1.plot(x, y1)
ax2.plot(x, y2)
plt.show()
```

### plt.subplots(nrows, ncols)関数

`fig`と `axes`が一度に生成できる。

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
print(type(fig))
print(type(axes))
print(type(axes[0]))

axes[0].plot(x, y1)
axes[1].plot(x, y2)
plt.show()
```

```python
fig, axes = plt.subplots(3, 4)
print(axes.shape)

axes[1, 2].plot(x, y2)
plt.show()
```

### ラベル

```python
fig, axes = plt.subplots(1, 2)

axes[0].set_title("plot title")
axes[0].set_xlabel("xlabel1")
axes[0].set_ylabel("xlabel2")
axes[0].set_xticks([-3, -2, -1, 3])
axes[0].set_yticks([0, 10, 20])
axes[0].plot(x, y1, label="something")
axes[0].legend()

axes[1].plot(x, y1)
axes[1].axis("off")

plt.show()
```

## 22 matplotlibを使ってグラフをPNG，PDFで保存する

[データサイエンスのためのPython入門22〜matplotlibを使ってグラフをPNG，PDFで保存する〜 - 米国データサイエンティストのブログ](https://datawokagaku.com/savefig_png_pdf/)

### グラフの大きさを調整

デフォルトの引数
- `figsize=(6.4, 4.8)`: 単位はインチ
- `dpi=100`

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x) * 2
```

まずはデフォルトのサイズと解像度

```python
fig, axes = plt.subplots(figsize=(6.4, 4.8), dpi=100)
axes.plot(x, y1)
plt.show()
```

縮小。目盛りも変わる。

```python
fig, axes = plt.subplots(figsize=(1, 1), dpi=100)
axes.plot(x, y1)
plt.show()
```

横長に拡大

```python
fig, axes = plt.subplots(figsize=(10, 3), dpi=100)
axes.plot(x, y1)
plt.show()
```

DPIを変更。目盛りは変わらないのに注目。

```python
fig, axes = plt.subplots(dpi=50)
axes.plot(x, y1)
```

```python
fig, axes = plt.subplots(dpi=120)
axes.plot(x, y1)
```

### pngで保存

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 3))

axes[0].set_title("first")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].plot(x, y1)

axes[1].set_title("second")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].plot(x, x**2)

fig.savefig("python-22-saving_sample.png")
```

ぎゅうぎゅうなので`tight_layout()`メソッドで調整する。

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 3))

axes[0].set_title("first")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].plot(x, y1)

axes[1].set_title("second")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].plot(x, x**2)

fig.tight_layout()
fig.savefig("python-22-saving_sample.png")
```

```python
fig, axes = plt.subplots()
axes.plot(x, y1, label="first")
axes.plot(x, y2, label="second")
axes.plot(x, y1 + y2, label="first + second")
axes.legend()
plt.show()
```

凡例の位置は変更できる。

```python
fig, axes = plt.subplots()
axes.plot(x, y1, label="first")
axes.plot(x, y2, label="second")
axes.plot(x, y1 + y2, label="first + second")
axes.legend(loc="center")
plt.show()
```

### pdfで保存する

```python
from matplotlib.backends.backend_pdf import PdfPages
```

```python
# pdfインスタンス作成。PDFファイルが出来る。
pdf = PdfPages("python-22-saving_sample.pdf")

# グラフ作成
fig, axes = plt.subplots()
axes.plot(x, y1, label="first")
axes.plot(x, y2, label="second")
axes.plot(x, y1 + y2, label="first + second")
axes.legend()

# PDFファイルに保存
pdf.savefig(fig)

# PDFファイルを閉じる
pdf.close()
```

```python
pdf = PdfPages("python-22-saving_multifig-sample.pdf")

for i in range(0, 10):
    fig, axes = plt.subplots()
    axes.set_title("ID: #{}".format(i))
    axes.plot(x, x**i)

    # グラフを保存、保存のたびに改ページされる
    pdf.savefig(fig)

pdf.close()
```

## 23 matplotlibで色々なグラフを描画する

[データサイエンスのためのPython入門23〜matplotlibで色々なグラフを描画する〜 - 米国データサイエンティストのブログ](https://datawokagaku.com/matplotlib_variousplots/)

### グラフの装飾

```python
x = np.linspace(-3, 3, 10)
y = np.exp(x)
```

まずはデフォルト

```python
fig, axes = plt.subplots()
axes.plot(x, y)
plt.show()
```

`Line2D`のプロパティを利用して装飾できる。

https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html

```python
fig, axes = plt.subplots()
axes.plot(
    x,
    y,
    color="red",
    lw=5,
    ls="--",
    marker="o",
    markersize=15,
    markerfacecolor="yellow",
    markeredgecolor="blue",
    markeredgewidth=4,
    alpha=0.5,
)
plt.show()
```

### 散布図：Axes.scatter()

```python
import pandas as pd

df = pd.read_csv("train.csv")
```

```python
fig, axes = plt.subplots()
axes.set_xlabel("Age")
axes.set_ylabel("Fare")
axes.scatter(df["Age"], df["Fare"], alpha=0.3)
plt.show()
```

### ヒストグラム：Axes.hist()

```python
fig, axes = plt.subplots()
axes.hist(df["Age"])
plt.show()
```

引数`bins`でヒストグラムの階級数が指定できる。

```python
fig, axes = plt.subplots()
axes.hist(df["Age"], bins=20)
plt.show()
```

**補足**: ここで文字列を指定すると`numpy.histogram_bin_edges`でサポートされた階級数の計算法が利用される。

[numpy.histogram_bin_edges — NumPy v2.4 Manual](https://numpy.org/doc/stable/reference/generated/numpy.histogram_bin_edges.html#numpy.histogram_bin_edges)

```python
fig, axes = plt.subplots()
axes.hist(df["Age"], bins="auto")
plt.show()
```

### 箱ひげ図 : Axes.boxplot()

```python
fig, axes = plt.subplots()
axes.boxplot(df["Age"].dropna())
plt.show()
```
